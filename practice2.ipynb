Dictionary

```python
House = {
    "wallColor": "beige",
    "plants": 10,
    "type": "colonial",
    "size": "medium"
}
```

```python
print(House["type"])
```

```python
House["bedrooms"] = 3
```

```python
print(House)
```

```python
House["wallColor"] = "white"
```

```python
print(House.values())
print(House.items())
```

```python
if "price" in House:
  print("house is on the market!")
else:
  print("house is not on the market")
```

```python
for key, value in House.items():
  print(key, "=>", value)
```

Problem: Make a phonebook

```python
Phonebook = {
    "Sarah": "869-372-0865",
    "Aiden": "858-382-4832",
    "Harper": "505-475-2933",
    "Julian": "432-543-2456"
}
```

```python
print("Hello! This is your phonebook: ")
for key, value, in Phonebook.items():
  if len(key) == 5:
    print(key, " |", value)
  else:
    print(key, "|", value)
```

```python
del Phonebook["Julian"]
```

```python
Phonebook["Julian"] = "309-298-4802"
```

```python
if "Julian" in Phonebook:
  print("Julian is here!")
else:
  print("Julian is not here")

print(Phonebook.get("Julian"))
```
